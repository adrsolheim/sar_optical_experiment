## Util Functions

```python
# sensor = ("opt"|"sar"|"lidar")
def save_tiff(img, meta, name, sensor="opt"):
    if sensor == "opt":
        filename = f'data/optical_Askvoll/{name}.tif'
    elif sensor == "sar":
        filename = f'data/SAR_Askvoll/SAR_ICEYE/{name}.tif'
    else:
        filename = f'data/lidar/{name}.tif'
    with rio.open(
        filename,
        'w',
        **meta
    ) as dst:
        dst.write(img, 1)

def calculate_ndvi(nir_band, red_band):
    return es.normalized_diff(nir_band, red_band)

def calculate_ndvi2(nir_band, red_band):
    return np.where(nir_band+red_band == 0., 0, (nir_band-red_band)/(nir_band+red_band))

def clip_to_aoi(mosaic: DatasetReader):
    with fiona.open('data/askvoll.geojson') as geo:
        aoi = [feature["geometry"] for feature in geo]
    clipped, transform = rio.mask.mask(mosaic, aoi, crop=True)
    meta = mosaic.meta.copy()
    meta.update({"transform": transform,
        "height":clipped.shape[1],
        "width":clipped.shape[2]})
    print(f'Saving clipped mosaic as {mosaic.name[:-4]}_clipped.tif')
    with rio.open(mosaic.name[:-4] + '_clipped.tif', 'w', **meta) as f:
        f.write(clipped)

#opt_3band = glob("data/optical_Askvoll/GeoEye-1/7-13-2019_Ortho_ColorBalance.tif")
#opt_3band_ds = rio.open(opt_3band[0])
#clip_to_aoi(opt_3band_ds)
#clip_to_aoi(sar_ds)
#chm = glob("data/lidar/mosaics/merged_chm_32632.tif")
#chm_ds = rio.open(chm[0])
#clip_to_aoi(chm_ds)
```

## Merge point cloud tiles
- Combine all `.las` patches into one file with LAStools: `lasmerge -i *.las -o merged.las`
- Reproject `EPSG:25832` -> `EPSG:32632` with LAStools: `las2las -i merged.las -target_epsg 32632 -o merged_reprojected.las`
    - "*SERIOUS WARNING: horizontal datum of source and target incompatible.*"
- Clip to AOI to reduce file size with LAStools: `lasclip -i merged_reprojected.las -olas -poly askvoll.shp`
    - "*WARNING: polygon 1 has 1 repeatd point(s) at count 8. ignoring ...*"
    - "*WARNING: polygon 1 has 1 repeatd point(s) at count 11. ignoring ...*"
    - "*WARNING: unlicensed. over 3 million points. output slightly distorted. tiny xyz noise. points permuted. intensity, gps_time, user_data and point_source ID zeroed.*"

    - *lascanopy*: `lascanopy -i *.las -merged -p 95 -step 0.4 -otif -o p95_merged.tif`
- **MeanH**
    - *Digital Terrain Modeler*: (DSM - DTM) => CHM
    - *lascanopy*: `lascanopy -i *.las -merged -avg -step 0.4 -otif -o meanh_merged.tif`
- **Dens**
    - *lascanopy*: `lascanopy -i *.las -step 0.4 -cov -dns -otif`
    - *`-dns` calculates the number of all points above the cover cutoff (e.g., `[1.3, 45)`) divided by the number of all returns*
- **Cover**
    - *lascanopy*: `lascanopy -i *.las -step 0.4 -cov -otif`
    - *`-cov` calculates the number of first returns above the cover cutoff (e.g., `[1.3, 45)`) divided by the number of all first returns*
- **Gini**

## Canopy Height Model (CHM)

**1.** Create a DSM from LiDAR (*Quick Terrain Modeler -> Import Model Data -> Export as GEOTIFF*)<br>
    - Grid sampling: `0.4` (optical resolution) <br>
**2.** Use DSM and DTM to get the Canopy Height Model (*Quick Terrain Modeler -> Subtract Model*)<br>
    - Merge tiles into mosaic, see `build_mosaic` <br>
**3.** Average pixels at correct resolution: `CHM -> numpy`<br>

<br style="clear:both" />

**4.** Create mask/shape from NDVI. Cut off high points that are "not green"

**95th percentile** might partially address miscalculated points (overhanging cliff issue)?


#### CHM alternatives
- Use `gdal2xyz` in QGIS to convert DTM to `.csv`
- Subtract DTM (ground) from point cloud with LAStools: `lasheight -i merged_reprojected_clipped.las -o height_point_cloud.las -replace_z -ground_points dtm.csv -parse ssxyz`
    - Merged point cloud too big: `ERROR: cannot alloc enough TIN triangles to triangulate 109943963 points.`
    - *NOTE: You CAN do this without DTM because the point cloud has ground points it can use*
    
    


```python
def build_mosaic(tile_files : list[str], name : str="merged_tiles"):
    x, y = rio.open(tile_files[0]).res
    vrt = gdal.BuildVRT("data/lidar/mosaics/merged_tiles.vrt", tile_files)
    gdal.Translate(f"data/lidar/mosaics/{name}.tif", vrt, xRes=x, yRes=y)
    vrt = None

#tile_files = glob("data/lidar/eksport_556468_20220210_models/NDH Askvoll 5pkt 2018/data/chm/*.tif")    
#build_mosaic(tile_files, "merged_chm")
    

def load_datasets():
    chm_files = glob("data/lidar/mosaics/merged_chm_clipped.tif")
    chm_ds = rio.open(chm_files[0])
    return chm_ds
    
def ndvi_mask(ndvi, threshold=0.3):
    return ndvi > threshold

def height_mask(chm, threshold=1.2):
    return chm >= 1.2
```

## Overhanging cliff issue

Overhanging cliffs prevents tree height calculation.
